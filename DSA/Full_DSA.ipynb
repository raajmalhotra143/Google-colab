#setup
this program handles writing, compiling, executing, and taking user input all in one step of all C++ codes

```python
from IPython.core.magic import register_cell_magic
import subprocess
import os

@register_cell_magic
def cpp(line, cell):
    # 1. Save the cell content to a C++ file
    filename = "solution.cpp"
    with open(filename, "w") as f:
        f.write(cell)

    # 2. Compile the C++ code
    compile_process = subprocess.run(["g++", filename, "-o", "solution"],
                                     capture_output=True, text=True)

    if compile_process.returncode != 0:
        print("‚ùå Compilation Failed!")
        print(compile_process.stderr)
        return
    else:
        print("‚úÖ Compilation Successful!")

    # 3. Ask for User Input (simulating interactive input)
    print("üëá Enter your input below (press Enter after pasting):")
    try:
        # We read lines until the user stops (useful for multi-line DSA inputs)
        # Note: In Colab, simple input() is often enough for single-line.
        # For multi-line, pasting typically works if the code handles whitespace.
        user_input = input()
    except EOFError:
        user_input = ""

    # 4. Run the executable and pipe the input
    print("\n--- RUNNING PROGRAM ---")
    print(f"Input provided to C++ program: {user_input}\n") # Added this line
    run_process = subprocess.run(["./solution"],
                                 input=user_input,
                                 capture_output=True,
                                 text=True)

    # 5. Display Output
    print(run_process.stdout)
    if run_process.stderr:
        print("Runtime Errors:", run_process.stderr)
```

```python
# sum of natural numbers
%%cpp
#include<iostream>
using namespace std;
int main(){
    int n, sum = 0 , sum2 = 0;
    cout << "Enter the value of n: ";
    cin >> n;
    for (int i = 0; i <= n; i++) {
        sum += i;

    }
    //can also be calculate using formula
    sum2 = n*(n+1) / 2;
    cout << "Sum is using loop: " << sum << endl;
    cout << "Sum is using formula : " << sum2 << endl;

    return 0;
}
```

The prompt 'Enter the value of n: is printed twice because the C++ program itself includes a cout << "Enter the value of n: "; statement which prints this message. Your input was provided through the Python input() function in the cpp magic command, which then gets passed to the C++ program's cin >> n; line. Even though the value is received, the cout statement within the C++ code still executes and displays its prompt.simply first program run only for giving input to the python then program again run with the provded input
