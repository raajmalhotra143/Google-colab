```python
# Install yt-dlp
!pip install yt-dlp

# Import library
from yt_dlp import YoutubeDL

# URL of the YouTube video
url = "https://youtu.be/6g6Q-dhOM0M?si=dGGril2oAQN4uxit"

# yt-dlp options for maximum quality (1080p if available)
ydl_opts = {
    'format': 'bestvideo[height<=1080]+bestaudio/best',
    'merge_output_format': 'mp4',
    'outtmpl': '%(title)s.%(ext)s'
}

# Download the video
with YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

print("Download complete!")
```

```python
# Install yt-dlp
!pip install yt-dlp

from yt_dlp import YoutubeDL
from google.colab import files
import os

# Your video URL
url = "https://youtu.be/6g6Q-dhOM0M?si=dGGril2oAQN4uxit"

# Download options
ydl_opts = {
    'format': 'bestvideo[height<=1080]+bestaudio/best',
    'merge_output_format': 'mp4',
    'outtmpl': 'video.mp4'
}

# Download video into Colab temp storage
with YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

print("Download complete! Now sending to your laptop...")

# Download directly to your device
files.download('video.mp4')

# OPTIONAL: delete from Colab after download
os.remove('video.mp4')
```
