```python
# Install yt-dlp
!pip install yt-dlp

from yt_dlp import YoutubeDL
from google.colab import files
import os

# ----------------------------
# USER SETTINGS
# ----------------------------

# 1) Enter YouTube video or playlist URL
url = "https://youtu.be/6g6Q-dhOM0M?si=dGGril2oAQN4uxit"

# 2) Choose resolution:
# Options: 720, 1080, 1440, 2160 (4K)
resolution = 2160

# 3) Rename file (no extension)
custom_name = "my_downloaded_video"

# 4) Is this a playlist? (True / False)
is_playlist = False  # Set True for playlist downloads

# 5) After saving to your laptop, delete from Colab?
auto_delete = True
# ----------------------------


# Build format string
format_selector = f"bestvideo[height<={resolution}]+bestaudio/best"

# Output filename
output_template = f"{custom_name}.%(ext)s"

# Downloader settings
ydl_opts = {
    'format': format_selector,
    'merge_output_format': 'mp4',
    'outtmpl': output_template,
    'noplaylist': not is_playlist   # True → download only single video
}

# Run downloader
with YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

print("\n=== DOWNLOAD COMPLETE IN COLAB ===\n")

# Handle single video download
if not is_playlist:
    final_name = custom_name + ".mp4"
    files.download(final_name)
    print("File sent to your laptop!")

    if auto_delete:
        os.remove(final_name)
        print("File removed from Colab.")

else:
    # Playlist: download all files one by one
    print("Playlist detected. Downloading each file to your laptop...")

    for file in os.listdir():
        if file.endswith(".mp4"):
            files.download(file)
            print(f"Downloaded → {file}")

            if auto_delete:
                os.remove(file)
                print(f"Removed from Colab → {file}")
```
